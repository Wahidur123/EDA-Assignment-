**11.Identify and remove outliers in the km_driven column using the IQR method.**

```python
import pandas as pd

file_path = 'BIKE DETAILS.csv'
bike_data = pd.read_csv(file_path)

# Calculate the IQR
Q1 = bike_data['km_driven'].quantile(0.25)
Q3 = bike_data['km_driven'].quantile(0.75)
IQR = Q3 - Q1


lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = bike_data[(bike_data['km_driven'] < lower_bound) | (bike_data['km_driven'] > upper_bound)]
print(f"Number of outliers in 'km_driven': {len(outliers)}")

# Remove outliers
bike_data_cleaned = bike_data[(bike_data['km_driven'] >= lower_bound) & (bike_data['km_driven'] <= upper_bound)]

print(f"Number of rows after removing outliers: {len(bike_data_cleaned)}")
```

**13.What is the average depreciation in selling price based on the bike's age (current year - manufacturing year)?**

```python
import pandas as pd
import datetime as dt

file_path = 'BIKE DETAILS.csv'
bike_data = pd.read_csv(file_path)

# Calculate the current year
current_year = dt.datetime.now().year

# Add a new column for bike age
bike_data['age'] = current_year - bike_data['year']

# Calculate the average selling price for each age group
average_depreciation = bike_data.groupby('age')['selling_price'].mean().reset_index()

# Display the results
print("Average selling price based on bike's age:")
print(average_depreciation)

# Visualize the depreciation trend
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(average_depreciation['age'], average_depreciation['selling_price'], marker='o', linestyle='-', color='b')

# Customize the plot
plt.title("Average Selling Price vs. Bike Age", fontsize=14)
plt.xlabel("Bike Age (Years)", fontsize=12)
plt.ylabel("Average Selling Price (INR)", fontsize=12)
plt.grid(True)
plt.show()
```

**14.Which bike names are priced significantly above the average price for their manufacturing year?**

```python
import pandas as pd

file_path = 'BIKE DETAILS.csv'
bike_data = pd.read_csv(file_path)

# Calculate the average price for each manufacturing year
year_avg_price = bike_data.groupby('year')['selling_price'].mean()

# Merge the average prices back into the original dataset
bike_data['avg_price_for_year'] = bike_data['year'].map(year_avg_price)

# Define a threshold for significantly above (e.g., 1.5x the average price for the year)
threshold_factor = 1.5
bike_data['is_significant'] = bike_data['selling_price'] > (bike_data['avg_price_for_year'] * threshold_factor)

# Filter bikes priced significantly above the average
significant_bikes = bike_data[bike_data['is_significant']]

# Display results
print("Bikes priced significantly above the average for their manufacturing year:")
print(significant_bikes[['name', 'year', 'selling_price', 'avg_price_for_year']])
```

**6.How does the average selling price of cars vary by customer gender and annual income?**

**15.How does the average car price vary across different car models and engine types?**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

car_sale_data = pd.read_csv('Car Sale.csv')

# Ensure relevant columns are numeric
car_sale_data['Engine'] = pd.to_numeric(car_sale_data['Engine'], errors='coerce')
car_sale_data['Price ($)'] = pd.to_numeric(car_sale_data['Price ($)'], errors='coerce')

# Drop rows with missing or invalid values in key columns
filtered_data = car_sale_data.dropna(subset=['Model', 'Engine', 'Price ($)'])

# Check if data exists after filtering
if filtered_data.empty:
    print("No valid data available for analysis.")
else:
    # Calculate average car price for each model and engine type
    average_price = filtered_data.groupby(['Model', 'Engine'])['Price ($)'].mean().reset_index()

    # Pivot table for visualization
    pivot_data = average_price.pivot(index='Model', columns='Engine', values='Price ($)')

    # Check if pivot table has valid data
    if pivot_data.empty:
        print("Pivot table is empty. No valid combinations of Model and Engine.")
    else:
        # Heatmap for visualization
        plt.figure(figsize=(14, 10))
        sns.heatmap(pivot_data, annot=True, fmt=".2f", cmap='YlGnBu', cbar=True)
        plt.title('Average Car Price Across Models and Engine Types', fontsize=16)
        plt.xlabel('Engine Type', fontsize=12)
        plt.ylabel('Car Model', fontsize=12)
        plt.xticks(rotation=45)
        plt.yticks(rotation=0)
        plt.tight_layout()
        plt.show()
```

**3.What is the distribution of discounted prices vs. actual prices?**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'amazon.csv'
amazon_data = pd.read_csv(file_path)

# Convert 'discounted_price' and 'actual_price' to numeric, handling errors
amazon_data['discounted_price'] = pd.to_numeric(amazon_data['discounted_price'], errors='coerce')
amazon_data['actual_price'] = pd.to_numeric(amazon_data['actual_price'], errors='coerce')

# Drop rows with NaN values in 'discounted_price' or 'actual_price'
amazon_data_cleaned = amazon_data.dropna(subset=['discounted_price', 'actual_price'])

# Plot a scatter plot for discounted vs. actual prices
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=amazon_data_cleaned,
    x='actual_price',
    y='discounted_price',
    alpha=0.7
)
plt.title('Discounted Prices vs. Actual Prices')
plt.xlabel('Actual Price (USD)')
plt.ylabel('Discounted Price (USD)')
plt.grid(True)
plt.show()

# Plot histograms for both discounted and actual prices
plt.figure(figsize=(12, 6))
sns.histplot(
    amazon_data_cleaned['actual_price'],
    label='Actual Price',
    kde=True,
    color='blue',
    alpha=0.6
)
sns.histplot(
    amazon_data_cleaned['discounted_price'],
    label='Discounted Price',
    kde=True,
    color='orange',
    alpha=0.6
)
plt.title('Distribution of Discounted Prices vs. Actual Prices')
plt.xlabel('Price (USD)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

**6.What are the most popular product keywords?**

```python
import pandas as pd
from collections import Counter

file_path = 'amazon.csv'
amazon_data = pd.read_csv(file_path)

# Extract 'product_name' and convert to lowercase
product_names = amazon_data['product_name'].dropna().str.lower()

# Split product names into words
words = product_names.str.split().explode()

# Count the word occurrences
word_counts = Counter(words)

# Get the top 10 most common keywords
top_keywords = word_counts.most_common(10)

# Display the results
print("Top 10 Most Popular Product Keywords:")
for keyword, count in top_keywords:
    print(f"{keyword}: {count}")
```

**9.Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot.**

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

spotify_data = pd.read_csv("spotify.csv")

# Select a subset of the dataset for top artists to avoid overcrowding the plot
top_artists = spotify_data['Artist'].value_counts().head(10).index
filtered_data = spotify_data[spotify_data['Artist'].isin(top_artists)]

# Box Plot
plt.figure(figsize=(14, 7))
sns.boxplot(data=filtered_data, x='Artist', y='Duration (ms)', palette="Set3")
plt.title("Box Plot of Track Duration by Top Artists", fontsize=16)
plt.xticks(rotation=45)
plt.xlabel("Artist")
plt.ylabel("Duration (ms)")
plt.show()

# Violin Plot
plt.figure(figsize=(14, 7))
sns.violinplot(data=filtered_data, x='Artist', y='Duration (ms)', palette="Set2", inner="quartile")
plt.title("Violin Plot of Track Duration by Top Artists", fontsize=16)
plt.xticks(rotation=45)
plt.xlabel("Artist")
plt.ylabel("Duration (ms)")
plt.show()
```
